**QUESTION** 1: List the Cities in which there is no flight from Moscow

```python
query = "Select distinct(city ->> 'en') as city \
	from airports \
	where (city ->> 'en') <> 'Moscow' \
	order by city" 

df = pd.read_sql_query(query, con = engine)
df.head(10)
```

**QUESTION 12**   Return all flight details in the indicated day 2017-08-28 include flight count ascending order,
departures count, time of departure and arrival time

```python
query = "Select flight_no, scheduled_departure :: time as dep_time, scheduled_arrival :: time as arrival_time, \
	departure_airport as departures, arrival_airport as arrival, count(flight_id) as No_of_flight \
	from flights \
	where scheduled_departure >= '2017-08-28'::date and scheduled_departure < '2017-08-29':: date \
	group by 1,2,3,4,5 \
	order by 4 asc " 

df = pd.read_sql_query(query, con = engine)
df.head(10)
```

**QUESTION 13**   Write a Query to arange the ange of model of aircrafts so short range is less than 2000. Middle range is more
than 2000 and less than 5000 & any range above 5000 is long range

```python
query = "Select (model ->> 'en') as model, case when range < 2000 then 'Short Range' \
							when range > 2000 and range < 5000 then 'Middle Range'\
							else 'Long_Range' end as Range \
	from aircrafts " 

df = pd.read_sql_query(query, con = engine)
df.head(10)
```

**QUESTION 14**  What is the shortest flight duration for each possible flight from Moscow to St. Petersburg, ann how
many tiems was the flight delayed for more than an hour`

```python
query = "Select f.flight_no, (f.scheduled_arrival  - f.scheduled_departure) as Scheduled_duration, \
		min(f.scheduled_arrival - f.scheduled_departure), max(f.scheduled_arrival - f.scheduled_departure), \
		Sum(Case when f.actual_departure > f.scheduled_departure + INTERVAL '1 hour' then 1 \
			Else 0  End) delays \
	from flights f \
	where(select city ->> 'en' from airports where airport_code = departure_airport) = 'Moscow' \
	and (select city ->> 'en' from airports where airport_code = arrival_airport) = 'St. Petersburg' \
	and f.status = 'Arrived' \
	Group by 1, (f.scheduled_arrival  - f.scheduled_departure); \
 " 

df = pd.read_sql_query(query, con = engine)
df.head(10)
```

**QUESTION 15**  Who travelled from Moscow(SVO) to Novosibirsk(OVB) on seat 1A , and when was the ticket booked?

```python
query = "SELECT t.passenger_name, b.book_date, bp.seat_no \
        FROM bookings b \
            JOIN tickets t \
                on b.book_ref = t.book_ref \
            JOIN boarding_passes bp \
                on t.ticket_no = bp.ticket_no \
            JOIN flights f \
                on bp.flight_id = f.flight_id \
        WHERE departure_airport = 'SVO' and arrival_airport = 'OVB' and bp.seat_no = '1A' \
            AND f.scheduled_departure::date = public.now()::date - INTERVAL '2 day'; \
         " 

df = pd.read_sql_query(query, con = engine)
df.head(10)
```

*QUESTION 16**  Find the most disciplined passengers who checked in frst for all thier flights. Take into account only those pssengers who took atleast two flights ?

```python
query = "SELECT t.passenger_name, t.ticket_no \
        FROM tickets t \
            JOIN boarding_passes bp \
                ON bp.ticket_no = t.ticket_no \
        GROUP BY t.passenger_name, t.ticket_no \
        HAVING MAX(bp.boarding_no) = 1 and count(*) > 1 \
         " 

df = pd.read_sql_query(query, con = engine)
df.head(10)
```
